##The `requests` Library

The `requests` library performs HTTP requests in Python with a clean, predictable interface. The same library used in the API module is used here. The pattern for retrieving a page is: construct a headers dictionary that always includes a `User-Agent`, call `requests.get()` with the URL and headers, check for HTTP errors with `response.raise_for_status()`, and access the response body via `response.text` for HTML or `response.json()` for JSON.

```python
import requests

HEADERS = {
    "User-Agent": "DataRollup-Scraping-Lab/1.0 "
                  "(Baruch College; contact: patrick.slattery@baruch.cuny.edu)"
}

response = requests.get("https://datarollup.io/associations/",
                        headers=HEADERS, timeout=10)
response.raise_for_status()
html = response.text
```



---



##Beautiful Soup (BS4) Essentials  

BeautifulSoup parses an HTML string into a navigable Python object. The four most common operations cover almost all introductory scraping needs: `find()` retrieves the first matching element, `find_all()` retrieves a list of all matches, attribute access via dictionary syntax retrieves attribute values, and `.get_text()` returns the visible text content with surrounding whitespace removed.

```python
from bs4 import BeautifulSoup

soup = BeautifulSoup(html, "html.parser")

# Find the first link
first_link = soup.find("a")
print(first_link.get("href"))
print(first_link.get_text())

# Find all links
all_links = soup.find_all("a")
for link in all_links:
    print(link.get("href"))
```

###Alternate: CSS Selectors  

The `.select()` method accepts CSS selector syntax for more concise queries. For students familiar with CSS, this is a cleaner path. For students who are not, `find_all()` with keyword arguments is the better starting point. Both are taught; CSS selectors are introduced as the second, more advanced syntax.

##Table Extraction

###Tables and `pandas.read_html()`

When the structured data on a page is presented in an HTML `<table>` element, the entire extraction can be performed in a single line: `pandas.read_html()` retrieves the HTML, identifies all `<table>` elements, and returns each as a Pandas DataFrame in a list. This is dramatically more concise than the manual BeautifulSoup approach when the data is genuinely tabular, and it is the right tool for the genuinely tabular case.

```python
import requests
import pandas as pd
from io import StringIO

HEADERS = {
    "User-Agent": "DataRollup-Scraping-Lab/1.0 "
                  "(Baruch College; contact: patrick.slattery@baruch.cuny.edu)"
}

response = requests.get("https://datarollup.io/associations/",
                        headers=HEADERS, timeout=10)
response.raise_for_status()
tables = pd.read_html(StringIO(response.text))
df = tables[0]
```
