### Exercise 2 — Class Instantiation & Methods  *(9:40–9:45)*

The `BankAccount` class below is already implemented.  
**Do not modify it.** Read it carefully, then complete the three lines below.

```python
# ── BankAccount class — provided, do not modify ──────────────────────────────

class BankAccount:
    """A simple bank account with a protected balance."""

    def __init__(self, owner, balance=0):
        self.owner    = owner
        self._balance = balance          # protected: use get_balance()

    def deposit(self, amount):
        """Add amount to the balance. Silently ignores non-positive amounts."""
        if amount > 0:
            self._balance += amount

    def withdraw(self, amount):
        """Deduct amount from the balance. Raises ValueError if insufficient funds."""
        if amount > self._balance:
            raise ValueError('NSF: insufficient funds')
        self._balance -= amount

    def get_balance(self):
        """Return the current balance."""
        return self._balance

    def __str__(self):
        return f'BankAccount({self.owner}, ${self._balance:.2f})'

    def __repr__(self):
        return f'BankAccount(owner={self.owner!r}, balance={self._balance})'

print('BankAccount class loaded.')
```

### Module 07: Database and SQLite  *(9:55–10:00)*

**Core concepts:**

| Layer | Key terms |
|:---|:---|
| Relational model | Table, row (record), column (attribute), primary key, foreign key, schema |
| DDL | `CREATE TABLE`, `ALTER TABLE`, `DROP TABLE` |
| DML | `INSERT INTO`, `UPDATE`, `DELETE FROM` |
| DQL | `SELECT … FROM … WHERE … GROUP BY … HAVING … ORDER BY` |
| Aggregates | `COUNT()`, `SUM()`, `AVG()`, `MAX()`, `MIN()` |
| Python | `sqlite3`, `conn.execute()`, cursor, `cursor.executescript()`, `COMMIT` |

**Critical distinctions:**

- **`WHERE`** filters individual rows *before* grouping — cannot reference aggregate expressions.
- **`HAVING`** filters groups *after* aggregation — can reference aggregate expressions.
- **`NULL`**: always test with `IS NULL` / `IS NOT NULL` — `= NULL` always returns zero rows.
- **`cursor.executescript()`** aborts silently on MySQL-specific statements (`CREATE DATABASE`, `USE`).

### Module 08: Unit Testing

**Core concepts:**

| Concept | Summary |
|:---|:---|
| Automated test | Written as code; runs in milliseconds; catches regressions |
| Unit test | Verifies one small, isolated behavior — isolated, repeatable, fast, self-describing |
| `unittest.TestCase` | Base class; `test_` prefix identifies test methods automatically |
| `setUp()` | Runs before every test method — creates a clean starting state |
| `tearDown()` | Runs after every test method — cleanup, even if the test failed |
| `assertEqual(a, b)` | Passes when `a == b` |
| `assertTrue(x)` | Passes when `bool(x)` is `True` |
| `assertFalse(x)` | Passes when `bool(x)` is `False` |
| `assertRaises(Exc, …)` | Passes when the specified exception is raised |
| `assertIsNone(x)` | Passes when `x is None` |
| TDD — Red | Write a failing test first — confirms the test actually tests something |
| TDD — Green | Write minimum code to make the test pass |
| TDD — Refactor | Clean up; passing tests confirm nothing broke |
| Regression | A previously working behavior that breaks due to a later code change |
| xUnit | Architecture shared by JUnit, NUnit, `unittest`: TestCase / TestSuite / TestRunner |
