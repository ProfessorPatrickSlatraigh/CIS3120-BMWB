## Lesson: The `main()` Function  
## *and* `if __name__ == "__main__":`     
  
In Python, it's common to organize code into functions to make it more modular and reusable. One key convention is the use of a `main()` function and the `if __name__ == "__main__":` construct. This lesson explains why these practices are important and how they work.

## Why Use a `main()` Function?

A `main()` function serves as the entry point for your Python program. Structuring your code this way offers several benefits:

1. **Code Organization**: Placing the main logic inside a function keeps the code organized. It allows you to break down the program into smaller, manageable functions.
2. **Reusability**: By separating the main logic from the rest of the code, you can easily import this script into other modules without automatically executing the code inside `main()`.
3. **Readability**: It makes the program easier to read by clearly defining where the execution of the program starts.

### Example:
```python
def main():
    print("This is the main function.")

if __name__ == "__main__":
    main()
```

The Role of `if __name__ == "__main__"`:
Python defines a special built-in variable called `__name__` for every module.   
When a Python file is run directly, the value of `__name__` is set to `"__main__"`. If the file is imported as a module in another script, `__name__` will be set to the name of the file.

Why Is This Useful?
This construct ensures that certain parts of the code are only executed when the file is run directly, not when it's imported. This is especially helpful for:

Testing: You can include test code that runs only when the script is executed directly.
Modularity: It allows your code to function both as a standalone program and as an importable module.
Example:
```
def greet():
    print("Hello, World!")

if __name__ == "__main__":
    greet()
```
In this case, the greet() function will only run if the file is executed directly. If it is imported into another script, the function will not automatically run.

## Lesson: Using f-Strings (formatted strings)  

In Python, f-strings (formatted string literals) are a convenient way to format strings by embedding expressions directly inside string literals. Introduced in Python 3.6, f-strings offer a readable and concise syntax for including variables or expressions in strings.

### Why Use f-Strings?

1. **Conciseness**: f-strings allow you to embed expressions inside curly braces `{}`, making your code shorter and more readable.
2. **Performance**: f-strings are generally faster than older methods like `str.format()` or concatenation using `+`.
3. **Readability**: With f-strings, you can directly see which variables are being embedded in the string, improving code clarity.

### Syntax

To create an f-string, place an `f` or `F` before the opening quote of the string, and then include variables or expressions inside curly braces `{}`.

#### Example:
```python
name = "Alice"
age = 30
print(f"My name is {name} and I am {age} years old.")
```

In this example, the values of name and age are embedded directly into the string.



`````````python
name = "Alice"
age = 30
print(f"My name is {name} and I am {age} years old.")
`````````

### Using Expressions in f-Strings  
f-strings can also evaluate expressions inside the curly braces, not just variables.

Example:
```
x = 10
y = 5
print(f"Sum of {x} and {y} is {x + y}.")
```
Output:
```
Sum of 10 and 5 is 15.
```
In this case, the expression {x + y} is evaluated, and the result is included in the string.

### Formatting with f-Strings  
f-strings support advanced formatting options, such as controlling the number of decimal places, alignment, and more.

Example:
```
price = 19.99
print(f"The price is ${price:.2f}")
```
Output:
```
The price is $19.99
```
Here, `:.2f` specifies that the number should be formatted with two decimal places.

### Multiline f-Strings
f-strings can be used with triple quotes for multiline strings, allowing you to format strings over multiple lines.

Example:
```
name = "Bob"
hobby = "painting"
print(f"""
Name: {name}
Hobby: {hobby}
""")
```
Output:
```
Name: Bob
Hobby: painting
```

## CodingGrace: Game04 - In a Room  

This level of the code defines an interactive text-based adventure game where the player chooses between entering a red or blue door. The `main()` function prompts the player for their name and begins the adventure by calling start_adventure(). In this function, the player picks either the red or blue door:

**Red Door**: Leads to a room with a red dragon. The player must choose to "flee" or stay. If they flee, the game restarts. If they stay, the dragon eats them.  

**Blue Door**: Leads to a room where nothing happens.  

The code uses conditional logic to determine the player's path based on their choices. If the input doesn't match "red" or "blue," an error message is displayed.  

`````````python
# Now we have a premise. We are in a room and we have two door to choose from.
# Each door leads to a room and we need to do something, in the red room
# specifically.

##### ROOMS #####
def blue_door_room():
    '''
    This is a blue door room.

    Nothing happens here, let's do one room at the time. :-)
    '''
    print("The door knob jiggles but nothing happens.")
    return

def red_door_room():
    '''
    The red door rooom contains a red dragon.

    If a player types "flee" as an answer, player returns to the room with
    two doors, otherwise the player dies.
    '''
    print("There you see a great red dragon.")
    print("It stares at you through one narrowed eye.")
    print("Do you flee for your life or stay?")

    next_move = input("> ")

    # Flee to return to the start of the game, in the room with the blue and
    # red door or die!
    if "flee" in next_move:
        start_adventure()
    else:
        print("It eats you. Well, that was tasty!")
### END ROOMS ###

def start_adventure():
    '''
    This function starts the adventure by allowing two options for
    players to choose from: red or blue door

    Chosen option will print out the door chosen.
    '''
    print("You enter a room, and you see a red door to your left and a blue door to your right.")
    door_picked = input("Do you pick the red door or blue door? > ")

    # Pick a door and we go to a room and something else happens
    if door_picked == "red":
        # This calls a function that contains stuff that happens in red_door_room
        red_door_room()
    elif door_picked == "blue":
        # This calls a function that contains stuff that happens in blue_door_room
        blue_door_room()
    else:
        print("Sorry, it's either 'red' or 'blue' as the answer. You're the weakest link, goodbye!")

def main():
    '''
    Gets the players name, print it out and starts the adventure.
    '''
    player_name =  input("What's your name? >")
    print(f"Your name is {player_name.upper()}")

    start_adventure()

if __name__ == '__main__':
    main()
`````````

## CodingGrace Game 9 - Using Dictionaries

This version of the text-based adventure game incorporates the use of dictionaries in Python.




The following changes are made to the action in the **Blue Door Room**:  
    
The player encounters a treasure chest and a guard.  

**Treasure Chest**:  
The player can open it to find items like diamonds, gold, silver, and a sword.
After finding the treasure, the player can take all the items (upgrading their sword) or leave them and focus on the guard.  
  
**Guard**:  
*The player can interact with the guard in three ways*:  
- **Attack**: Leads to the player’s death, ending the game.
- **Check**: Observes that the guard is still asleep but offers no further action.
- **Sneak**: Successfully sneaks past the guard to freedom, completing the game.

`````````python
# Now we have a premise. We are in a room and we have two door to choose from.
# We are still in the blue room. Now we have to deal with the guard
# after taking or leaving the treasure chest.
#
# Introducing dictionaries and while loop

##### ACTIONS #####
def you_died(why):
    '''
    In: Passing in the string showing player how they dies

    Result:
    Prints reason why they player died.
    Programme exits without error.
    '''
    print(f"{why}. Good job!")
    # This exits the program entirely.
    exit(0)

### END ACTIONS ###

### CHARACTERS ###
def guard():
    '''
    Encountering the guard, the player chooses to attack, check or sneak.
    - attack: player dies and it is game over
    - check: sees what the guard is doing, but nothing else happens, and get 3 options again
    - sneak: player sneaks past the guard and wins the game
    '''
    # Actions on the guard
    actions_dict = {"check":"You see the guard is still sleeping, you need to get to that door on the right of him. What are you waiting for?",
                    "sneak":"You approach the guard, he's still sleeping. Reaching for the door, you open it slowly and slip out.",
                    "attack":"You swiftly run towards the sleeping guard and knock him out with the hilt of your shiney sword. Unfortunately it wasn't hard enough."}

    # While loop
    while True:
        action = input("What do you do? [attack | check | sneak] >").lower()
        if action in actions_dict.keys():
            print(actions_dict[action])
            if action == "sneak":
                print("You just slipped through the door before the guard realised it.")
                print("You are now outside, home free! Huzzah!\n")
                return
            elif action == "attack":
                you_died("Guard woke with a grunt, and reached for his dagger and before you know it, the world goes dark and you just died. \n<GAME OVER>")

### END CHARACTERS ###

##### ROOMS #####
def blue_door_room():
    '''
    The player finds a treasure chest, options to investigate the treasure chest or guard.

    If player chooses
    - Treasure chest: show its contents; option to take treasure or ignore it (proceeds to guard)
    - Guard: After checking treasure chest, ignoring treasure chest to check guard, it calls guard() function
    '''
    # So our treasure_chest list contains 4 items.
    treasure_chest = ["diamonds", "gold", "silver", "sword"]
    print("You see a room with a wooden treasure chest on the left, and a sleeping guard on the right in front of the door")

    # Ask player what to do.
    action = input("What do you do? > ")

    # This is a way to see if the text typed by player is in the list
    if action.lower() in ["treasure", "chest", "left"]:
        print("Oooh, treasure!")

        print("Open it? Press '1'")
        print("Leave it alone. Press '2'")
        choice = input("> ")

        if choice == "1":
            print("Let's see what's in here... /grins")
            print("The chest creaks open, and the guard is still sleeping. That's one heavy sleeper!")
            print("You find some")

            # for each treasure (variable created on the fly in the for loop)
            # in the treasure_chest list, print the treasure.
            for treasure in treasure_chest:
                print(treasure)

            # So much treasure, what to do? Take it or leave it.
            print("What do you want to do?")
            # Get number of items in treasure chest with len))
            num_items_in_chest = len(treasure_chest)

            print(f"Take all {num_items_in_chest} treasure, press '1'")
            print("Leave it, press '2'")

            treasure_choice = input("> ")
            if treasure_choice == "1":
                treasure_chest.remove("sword")
                print("\tYou take the shinier sword from the treasure chest. It does looks exceedingly shiney.")
                print("\tWoohoo! Bounty and a shiney new sword. /drops your crappy sword in the empty treasure chest.")

                temp_treasure_list = treasure_chest[:]
                treasure_contents = ", ".join(treasure_chest)
                print(f"\tYou also receive {treasure_contents}.")

                # Removing all the rest of the items in the treasure chest
                for treasure in temp_treasure_list:
                    # Use list remove() function to remove each item in the chest.
                    treasure_chest.remove(treasure)

                # Add the old sword in place of the new sword
                treasure_chest.append("crappy sword")
                print(f"\tYou close the lid of the chest containing {treasure_chest} for the next adventurer. /grins")
                print("Now onward to get past this sleeping guard and the door to freedom.")
            elif treasure_choice == "2":
                print("It will still be here (I hope), right after I get past this guard")
        elif choice == "2":
            print("Who needs treasure, let's get out of here.'")
    elif action.lower() in ["guard", "right"]:
        print("Let's have fun with the guard.")
    else:
        print("Well, not sure what you picked there, let's poke the guard cos it's fun!")
    guard()

def red_door_room():
    '''
    The red door rooom contains a red dragon.

    If a player types "flee" as an answer, player returns to the room with two doors,
    otherwise the player dies.
    '''
    print("There you see a great red dragon.")
    print("It stares at you through one narrowed eye.")
    print("Do you flee for your life or stay?")

    next_move = input("> ")

    # Flee to return to the start of the game, in the room with the blue and red door or die!
    if "flee" in next_move:
        start_adventure()
    else:
        # You call the function you_died and pass the reason why you died as
        # a string as an argument.
        you_died("It eats you. Well, that was tasty!")
### END ROOMS ###

def start_adventure():
    '''
    This function starts the adventure by allowing two options for
    players to choose from: red or blue door

    Chosen option will print out the door chosen.
    '''
    print("You enter a room, and you see a red door to your left and a blue door to your right.")
    door_picked = input("Do you pick the red door or blue door? > ")

    # Pick a door and we go to a room and something else happens
    if door_picked == "red":
        red_door_room()
    elif door_picked == "blue":
        blue_door_room()
    else:
        print("Sorry, it's either 'red' or 'blue' as the answer. You're the weakest link, goodbye!")

def main():
    '''
    Gets the players name, print it out and starts the adventure.
    '''
    player_name =  input("What's your name? >")
    print(f"Your name is {player_name.upper()}")

    start_adventure()

if __name__ == '__main__':
    main()
`````````

### Include `rps()` Function  

We can include the defintion of the `rps()` function that plays **Rock, Paper, Scissors**.  


`````````python
import random  # Don't forget that we need the `random` library

def rps(user_wins=True):
    """Plays Rock, Paper, Scissors based on user's choice.
    Argument toggles whether user wins or loses (Boolean)
    or uses weighted probabilities if a list of three floats summing to 1.0.
    Returns a tuple in the format ("You typed", user_choice, result_message).
    Accepts 'no thanks' or 'done' to exit without playing.
    """

    # Define constants for choices
    CHOICES = ["Rock", "Paper", "Scissors"]
    WIN_SCENARIOS = {
        "Rock": "Scissors",
        "Paper": "Rock",
        "Scissors": "Paper"
    }

    # Define exit options
    EXIT_OPTIONS = ["No thanks", "Done"]

    # Prompt user for input
    print("Let's play Rock, Paper, Scissors!")
    prompt = "Choose rock, paper, or scissors by entering the word (or type 'no thanks' or 'done' to exit): "
    user_input = input(prompt).strip().capitalize()

    # Check if user wants to exit
    if user_input in EXIT_OPTIONS:
        print("Maybe next time! Exiting the game.")
        return "You chose not to play", user_input, "Game exited"

    # Validate user input
    if user_input not in CHOICES:
        print("Invalid choice. Please enter Rock, Paper, or Scissors.")
        return "Invalid choice", user_input, "Invalid input"

    # Check if user_wins is a list of probabilities that sum to 1.0
    if isinstance(user_wins, list) and len(user_wins) == 3 and all(isinstance(w, float) for w in user_wins) and sum(user_wins) == 1.0:
        # Use random.choices with the provided weights
        computer_choice = random.choices(CHOICES, weights=user_wins, k=1)[0]
        if WIN_SCENARIOS[user_input] == computer_choice:
            result_message = "You win!"
        elif WIN_SCENARIOS[computer_choice] == user_input:
            result_message = "You lose!"
        else:
            result_message = "It's a tie!"

    else:
        if user_wins:
            computer_choice = WIN_SCENARIOS[user_input]
            result_message = "You win!"
        else:
            computer_choice = next(choice for choice, beats in WIN_SCENARIOS.items() if beats == user_input)
            result_message = "You lose!"

    # Output results
    print(f"{computer_choice}. {result_message}")
    return "You typed", user_input, result_message
`````````

## CodingGrace NextLevel B - Adding a Global Dictionary for User Info  
  


We can further enrich the text-based adventure game by using a **dictionary** as a **global variable** that can be read and updated from any function in the game.  We will create a `user_info` dictionary and store the following information about the player in it:  
  
- `name` - a string  
- `level` - an integer  
- `inventory` - a list
- `location` - a string  
- `health` - an integer
- `choices` - a list  




### How Global Variables Work  
  

<b>1) Scope:</b>  
  
- A global variable is defined outside any function, making it accessible to all functions in the program.  
- To modify a global variable inside a function, you must explicitly declare it as global.  
  
<b>2) Persistence</b>:  
  
- Global variables persist throughout the runtime of the program, allowing functions to access and update the same data.  

<b>3) Shared Access</b>:  
  
- Multiple functions can share and update the same global variable, enabling centralized control over data.  


### Why Use **Global Variables** In Games?  
  

<b>1) Centralized State Management</b>:  
  
- In a text-based game, global variables can store player information (e.g., name, inventory, health) that needs to be accessed and updated across different rooms or scenarios.  
  
<b>2) Ease of Data Sharing</b>:  
  
- Functions like `red_room()`, `blue_room()`, and `green_room()` can all access and update a single `player_info` dictionary, ensuring the player's state remains consistent.  
  
<b>3) Simplified Development</b>:  
  
- Using global variables reduces the complexity of passing the same data between functions repeatedly.  
   

### Example of **Global Variable** Usage  
  

Here's a simple example of how global variables work in a game setting:  
  


`````````python
# Define a global dictionary to store player information
player_info = {
    "name": "",
    "level": 1,
    "inventory": [],
    "location": "Starting Room",
    "health": 100,
    "choices": []
}

def main():
    print("Welcome to the game!")
    set_player_name()
    show_player_info()

def set_player_name():
    """Sets the player's name."""
    global player_info  # Access the global variable
    player_info["name"] = input("Enter your player name: ").strip()

def show_player_info():
    """Displays the current player information."""
    global player_info  # Access the global variable
    print("\nPlayer Information:")
    for key, value in player_info.items():
        print(f"{key.capitalize()}: {value}")

def main():
    print("Welcome to the game!")
    set_player_name()
    show_player_info()

if __name__ == "__main__":
    main()
`````````

### Best Practices for **Global Variables**  
  

<b>1) Use Sparingly</b>:  
  
- Overusing global variables can lead to hard-to-debug code. Use them only when necessary.  
  
<b>2) Organize Data</b>:  
  
- For complex programs, store related data in global dictionaries or classes to keep it manageable.  
  
<b>3) Document Clearly</b>:   
  
 - Always document global variables and their purpose to make the code easier to understand and maintain.

`````````python
##### IMPORT LIBRARIES #####
import os
# We add an import statement to access the random library
import random  # We need the `random` library for our rps() function
##### END IMPORT LIBRARIES #####


##### INITIALIZE VARIABLES #####
# We add a section with the initialization of player_info variables
# in a dictionary to store player information
player_info = {
    "name": "",
    "level": 1,
    "inventory": [],
    "location": "Starting Room",
    "health" : 100,
    "choices": []
}

##### END INITIALIZE VARIABLES #####


##### ACTIONS #####

def you_died(why):
    print_game_over()
    # You expect a reason why the player died. It's a string.
    print("{}. Too bad!".format(why))

    # This exits the program entirely.
    exit(0)

# We add a you_won() function for our Green Room and rps() game
def you_won(how):
    print_game_over()
    # You expect an explanation how the player won. It's a string.
    print("{}. The game is over. You won!".format(how))

    # This exits the program entirely.
    exit(0)

# We add a next_level() functiom to increase the player's level in the game
def next_level(player_info_arg):
    """Simulates progression to the next level."""
    #  player_info_arg accesses the player_info dictionary argument passed
    player_info_arg["level"] += 1
    print("\nCongratulations! You have advanced to Level", player_info["level"])
    return player_info_arg  # return the updated player_info dictionary


def rps(user_wins=True):
    """Plays Rock, Paper, Scissors based on user's choice.
    Argument toggles whether user wins or loses (Boolean)
    or uses weighted probabilities if a list of three floats summing to 1.0.
    Returns a tuple in the format ("You typed", user_choice, result_message).
    Accepts 'no thanks' or 'done' to exit without playing.
    """

    # Define constants for choices
    CHOICES = ["Rock", "Paper", "Scissors"]
    WIN_SCENARIOS = {
        "Rock": "Scissors",
        "Paper": "Rock",
        "Scissors": "Paper"
    }

    # Define exit options
    EXIT_OPTIONS = ["No thanks", "Done"]

    # Prompt user for input
    print("Let's play Rock, Paper, Scissors!")
    prompt = "Choose rock, paper, or scissors by entering the word (or type 'no thanks' or 'done' to exit): "
    user_input = input(prompt).strip().capitalize()

    # Check if user wants to exit
    if user_input in EXIT_OPTIONS:
        print("Maybe next time! Exiting the game.")
        return "You chose not to play", user_input, "Game exited"

    # Validate user input
    if user_input not in CHOICES:
        print("Invalid choice. Please enter Rock, Paper, or Scissors.")
        return "Invalid choice", user_input, "Invalid input"

    # Check if user_wins is a list of probabilities that sum to 1.0
    if isinstance(user_wins, list) and len(user_wins) == 3 and all(isinstance(w, float) for w in user_wins) and sum(user_wins) == 1.0:
        # Use random.choices with the provided weights
        computer_choice = random.choices(CHOICES, weights=user_wins, k=1)[0]
        if WIN_SCENARIOS[user_input] == computer_choice:
            result_message = "You win!"
        elif WIN_SCENARIOS[computer_choice] == user_input:
            result_message = "You lose!"
        else:
            result_message = "It's a tie!"

    else:
        if user_wins:
            computer_choice = WIN_SCENARIOS[user_input]
            result_message = "You win!"
        else:
            computer_choice = next(choice for choice, beats in WIN_SCENARIOS.items() if beats == user_input)
            result_message = "You lose!"

    # Output results
    print(f"{computer_choice}. {result_message}")
    return "You typed", user_input, result_message

##### END ACTIONS #####


##### CHARACTERS #####

def guard():
    print_guard()
    # The guard
    print("You approach the guard, he's still sleeping.")
    print("Suddenly you knocked a wooden cask with a mug on it... CRASSH!")
    print("\n'Hey, you! What you doing 'ere?'")

    # Guard is not moving initially
    guard_moved = False

    # - When a player dies, it calls you_died() and it exits() the program.
    # - When a player escapes through the door, you return to the previous function which
    #   called this function.
    while True:
        next_action = input("[run | door] > ").lower()
        if next_action == "run" and guard_moved:
            you_died("The guard was faster than he looks and your world goes dark...")
        elif next_action == "run" and not guard_moved:
            print("Guard jumps up and looks the other way, missing you entirely.")
            guard_moved = True
        elif next_action == "door" and guard_moved:
            print("You just slipped through the door before the guard realised it.")
            print("You are now outside, home free! Congratulations!")
            return
        elif next_action == "door" and not guard_moved:
            you_died("The guard was faster than he looks and your world goes dark...")
        else:
            print("Not sure what you meant there... try again.")

##### END CHARACTERS #####


##### ROOMS #####

def blissful_ignorance_of_illusion_room(player_info_arg):
    """Handles player interactions in the Blue Room (Blissful Ignorance of Illusion)."""
    #  player_info_arg accesses the player_info dictionary argument passed
    print_chest()
    print("\nYou have entered the Blue Room.")

    # Update player location
    player_info_arg["location"] = "Blue Room"

    # Player gains health but loses time
    healing = 30
    illusion = "Crystal Mirror"
    player_info_arg["health"] += healing
    if illusion not in player_info_arg["inventory"]:
        player_info_arg["inventory"].append(illusion)
        print(f"The blissful ignorance restores {healing} health and grants you the {illusion}.")

    # Log the choice
    player_info_arg["choices"].append("Blue Room")

    # Show updated player information
    print("\nPlayer Info:")
    for key, value in player_info_arg.items():
        print(f"{key.capitalize()}: {value}")

    # A list may be empty as well.
    # So our treasure_chest list contains 4 items.
    treasure_chest = ["diamonds", "gold", "silver", "sword"]
    print("You see a room with a wooden treasure chest on the left, and a sleeping guard on the right in front of the door.")

    # Ask player what to do.
    action = input("\nWhat do you do? > ")

    # This is a way to see if the text typed by player is in the list
    if action.lower() in ["treasure", "chest", "left"]:
        print("Oh, it's full of treasure!")

        print("Open it?  Press '1'")
        print("Leave it alone.  Press '2'")
        choice = input("> ")

        # Try just leaving 1 and 2 as a number
        # Change to string and see what happens
        if choice == "1":
            print("Let's see what's in here... /grins")
            print("The chest creaks open, and the guard is still sleeping. That's one heavy sleeper!")
            print("You find: {}.".format(", ".join(treasure_chest)))

            # *** THE FOLLOWING ORIGINAL CODE ITERATED THROUGH TREASURE ***
            # Get number of items in treasure chest with len))
            # num_items_in_chest = len(treasure_chest)
            # print("You find some ", end='')

            # for each treasure (variable created on the fly in the for loop)
            # in the treasure_chest list, print the treasure.
            # for treasure in treasure_chest:
            #    print(treasure)

            # So much treasure, what to do? Take it or leave it.
            print("\nWhat do you want to do?")
            print("Take all {} treasure, press '1'".format(len(treasure_chest)))
            print("Leave it, press '2'")

            treasure_choice = input("> ")
            if treasure_choice == "1":
                print("\tAmazing! Bounty and a shiney new sword. /drops your crappy sword in the empty treasure chest.")
                print("\tYou just received [{}]".format(", ".join(treasure_chest)))
            elif treasure_choice == "2":
                print("It will still be here (I hope), right after I get past this guard.")

            # Picked up treasure or left it, you will now encounter the guard.
            # Let's call the guard() function here.
            guard()
            return player_info_arg  # return the updated player_info dictionary
    else:
        # Let's call the guard() function here as well, no point writing a bunch of same code
        # twice (or more). It's good to be able to re-use code.
        print("The guard is more interesting, let's go that way!")
        guard()
        return player_info_arg  # return the updated player_info dictionary


def painful_truth_of_reality_room(player_info_arg):
    """Handles player interactions in the Red Room (Painful Truth of Reality)."""
    #  player_info_arg accesses the player_info dictionary argument passed
    print_monster()
    print("\nYou have entered the Red Room.")

    # Update player location
    player_info_arg["location"] = "Red Room"

    # Player loses health but gains knowledge
    damage = 20
    knowledge = "Truth Scroll"
    player_info_arg["health"] -= damage
    if knowledge not in player_info_arg["inventory"]:
        player_info_arg["inventory"].append(knowledge)
        print(f"The painful truth costs you {damage} health but grants you the {knowledge}.")

    # Log the choice
    player_info_arg["choices"].append("Red Room")

    # Show updated player information
    print("\nPlayer Info:")
    for key, value in player_info_arg.items():
        print(f"{key.capitalize()}: {value}")

    print("You see the great evil Melon Usk.")
    print("He, it, whatever stares at you and you go insane.")
    print("Do you flee for your life or eat your head?")

    next_move = input("> ")

    # Flee to return to the start of the game, in the room with the blue and red door or die!
    if "flee" in next_move:
        start_new_adventure(player_info_arg)
    else:
        # You call the function you_died and pass the reason why you died as
        # a string as an argument.
        you_died("You died. Well, that was tasty!")
        return player_info_arg  # return the updated player_info dictionary (just in case)


def green_magic_room(player_info_arg):
    """Handles player interactions in the Green Room."""
    #  player_info_arg accesses the player_info dictionary argument passed
    print_magician()
    print("Welcome to the green magic room.")
    print("Prepare yourself to play a magic game of Rock, Paper, Scissors.")
    print("If you win, you will receive a gold coin.")
    print("If you do not win, you will return to the start of the game.\n")
    print

    # Update player location
    player_info_arg["location"] = "Green Room"

    # Add a special item to the player's inventory
    special_item = "Emerald Amulet"
    if special_item not in player_info_arg["inventory"]:
        player_info_arg["inventory"].append(special_item)
        print(f"You found a {special_item} and added it to your inventory!")

    # Show updated player information
    print("\nPlayer Info:")
    for key, value in player_info_arg.items():
        print(f"{key.capitalize()}: {value}\n")


    rps_result = [None, None, "tie"]    # setting up to play until not tied
    while 'tie' in rps_result[2]:       # playing while not tied
        rps_result = rps([.3,.4,.3])    # playing Rock, Paper, Scissors
        if "tie" not in rps_result[2]:  # testing if not tied
            break                       # leaving the game if not tied
        print("You tied and can play Rock, Paper, Scissors again.\n")

    # You did not win Rock, Paper, Scissors and must start the game over!
    if not "win" in rps_result[2]:
        start_new_adventure(player_info_arg)
    else:
        # You won the game of Rock, Paper, Scissors and get a gold coin
        # and can leave the game.
        you_won("The magician gave you a gold coin. Congratulations!")
        return player_info_arg  # return the updated player_info dictionary (just in case)


##### END ROOMS #####


##### CONTROL FUNCTIONS #####

def get_player_name(player_info_arg):
    # LOCAL VARIABLES
    #  player_info_arg accesses the player_info dictionary argument passed
    # The player enters their name and gets assigned to the key "name" in player_info
    player_info_arg["name"] = input("Enter your player name: ").strip()

    # This is just an alternative name that the game wants to call the player
    alt_name = "Rainbow Unicorn"
    answer = input("Your name is {}, is that correct? [Y|N] > ".format(alt_name.upper()))
    if answer.lower() in ["y", "yes"]:
        player_info_arg["name"] = alt_name
        print("You are fun, {}! Let's begin our adventure!".format(player_info["name"].upper()))
    elif answer.lower() in ["n", "no"]:
        print("Ok, picky. {} it is. Let's get started on our adventure.".format(player_info["name"].upper()))
    else:
        print("Trying to be funny? Well, you will now be called {} anyway.".format(alt_name.upper()))
        player_info_arg["name"] = alt_name
        print_smiley_face()

    # Now notice that we are returning the variable called player_info["name"].
    # In main(), this is now redundant because it is a global variable.
    # This is why indentation is important, variables declared in this block only exists in the block
    return player_info_arg

def start_new_adventure(player_info_arg):
    #  player_info_arg accesses the player_info dictionary argument passed
    print_new_dungeon()
    print("You enter a room, and you see a red door to your left and blue and green doors to your right.")
    door_picked = input("Do you pick the red door, blue door, or green door? > ")

    # Pick a door and we go to a room and something else happens
    if door_picked[:3].lower() == "red":
        player_info_arg = painful_truth_of_reality_room(player_info_arg)
    elif door_picked[:4].lower() == "blue":
        player_info_arg = blissful_ignorance_of_illusion_room(player_info_arg)
    elif door_picked[:5].lower() == "green":
        player_info_arg = green_magic_room(player_info_arg)
    else:
        print("Sorry, it's either 'red', 'blue', or 'green' as the answer. You're the weakest link, goodbye!")
        return player_info_arg  # return the updated player_info dictionary

def main(player_info_main):
    # os.system("clear")
    #  player_info_main accesses the player_info dictionary argument passed
    print("\nWelcome to the game!")

    # Calls get_player_name and returns the player name
    player_info_main =  get_player_name(player_info_main)

    start_new_adventure(player_info_main) # this starts the game with Red, Blue, and Green rooms

    print("\nThe end\n")
    print("Thanks for playing, {}".format(player_info_main["name"].upper()))  # revised to use the dictionay attribute

##### END CONTROL FUNCTIONS #####


##### TEXT IMAGE FUNCTIONS ######

def print_monster():
    print()
    print("                           |                     | ")
    print("                        \     /               \     / ")
    print("                       -= .'> =-             -= <'. =- ")
    print("                          '.'.                 .'.' ")
    print("                            '.'.             .'.' ")
    print("                              '.'.----^----.'.' ")
    print("                               /'==========='\ ")
    print("                           .  /  .-.     .-.  \  . ")
    print("                           :'.\ '.O.') ('.O.' /.':   ")
    print("                           '. |               | .'   ")
    print("                             '|      / \      |' ")
    print("                              \     (o'o)     / ")
    print("                              |\             /| ")
    print("                              \('._________.')/ ")
    print("                               '. \/|_|_|\/ .'                ")
    print("                                /'._______.'\ -- ")
    print()

def print_chest():
    print()
    print("                      _.--. ")
    print("                  _.-'_:-'|| ")
    print("              _.-'_.-::::'|| ")
    print("         _.-:'_.-::::::'  || ")
    print("       .'`-.-:::::::'     || ")
    print("      /.'`;|:::::::'      ||_ ")
    print("     ||   ||::::::'     _.;._'-._ ")
    print("     ||   ||:::::'  _.-!oo @.!-._'-. ")
    print("     ('.  ||:::::.-!()oo @!()@.-'_.| ")
    print("      '.'-;|:.-'.&$@.& ()$%-'o.'-U|| ")
    print("        `>'-.!@%()@'@_%-'_.-o _.|'|| ")
    print("         ||-._'-.@.-'_.-' _.-o  |'|| ")
    print("         ||=[ '-._.-+U/.-'    o |'|| ")
    print("         || '-.]=|| |'|      o  |'|| ")
    print("         ||      || |'|        _| '; ")
    print("         ||      || |'|    _.-'_.-' ")
    print("         |'-._   || |'|_.-'_.-' ")
    print("          '-._'-.|| |' `_.-' ")
    print("              '-.||_/.-' ")
    print()

def print_guard():
    print()
    print("                                                  ___I___ ")
    print("                                                 /=  |  #\ ")
    print("                                                /.__-| __ \ ")
    print("                                                |/ _\_/_ \| ")
    print("                                                (( __ \__)) ")
    print("                                             __ ((()))))()) __ ")
    print("                                           ,'  |()))))(((()|# `. ")
    print("                                          /    |^))()))))(^|   =\ ")
    print("                                         /    /^v^(())()()v^;'  .\ ")
    print("                                         |__.'^v^v^))))))^v^v`.__| ")
    print("                                        /_ ' \______(()_____(   | ")
    print("                                   _..-'   _//_____[xxx]_____\.-| ")
    print("                                  /,_#\.=-' /v^v^v^v^v^v^v^v^| _| ")
    print("                                  \)|)      v^v^v^v^v^v^v^v^v| _| ")
    print("                                   ||       :v^v^v^v^v^v`.-' |#  \, ")
    print("                                   ||       v^v^v^v`_/\__,--.|\_=_/ ")
    print("                                   ><       :v^v____|  \_____|_ ")
    print("                                ,  ||       v^      /  \       / ")
    print("                               //\_||_)\    `/_..-._\   )_...__\ ")
    print("                              ||   \/  #|     |_='_(     |  =_(_ ")
    print("                              ||  _/\_  |    /     =\    /  '  =\ ")
    print("                               \\\/ \/ )/    |=____#|    '=....#| ")
    print()

def print_game_over():
    print()
    print("   _____          __  __ ______    ______      ________ _____  ")
    print("  / ____|   /\   |  \/  |  ____|  / __ \ \    / /  ____|  __ \ ")
    print(" | |  __   /  \  | \  / | |__    | |  | \ \  / /| |__  | |__) |")
    print(" | | |_ | / /\ \ | |\/| |  __|   | |  | |\ \/ / |  __| |  _  / ")
    print(" | |__| |/ ____ \| |  | | |____  | |__| | \  /  | |____| | \ \ ")
    print("  \_____/_/    \_\_|  |_|______|  \____/   \/   |______|_|  \_\\")
    print()

def print_smiley_face():
    print()
    print("       *****       ")
    print("    **       **    ")
    print("  **  O   O   **   ")
    print(" **     \\_/     **  ")
    print(" **              **  ")
    print("  **   \\___/   **   ")
    print("    **       **    ")
    print("       *****       ")
    print()

def print_magician():
    print()
    print("                                                  _____ ")
    print("                                                 /     \\ ")
    print("                                                /       \\ ")
    print("                                               /_________\\ ")
    print("                                              |         | ")
    print("                                              |  () ()  | ")
    print("                                               \\   ^   / ")
    print("                                                \\_____/ ")
    print("                                                 ||||| ")
    print("                                                 ||||| ")
    print("                                             ____/  _  \\____ ")
    print("                                            /    |       |   \\ ")
    print("                                           /     |       |    \\ ")
    print("                                          |      |       |     | ")
    print("                                           \\_____|_______|_____/ ")
    print("                                             /   _______   \\ ")
    print("                                            /               \\ ")
    print("                                           |    O     O     | ")
    print("                                            \\_______________/ ")
    print("                                             |||         ||| ")
    print("                                             |||         ||| ")
    print("                                             |||         ||| ")
    print("                                            (___)       (___) ")
    print()

def print_new_dungeon():
    print()
    print("   _____________________________________________________________________________")
    print(" /|     -_-                                                           _-      |\ ")
    print("/ |_-_- _                                                     -_- _-   -_-   -| \   ")
    print("  |                                  _-  _--                                     | ")
    print("  |                                  ,                                           |")
    print("  |      .-'  '-.        '(        .-'  '-.       '(        .-'  '-.            |")
    print("  |    . |        .      )'      . |        .    )'      . |        .          |")
    print("  |   /   |   ()    \      U      /   |   ()    \      U      /   |   ()    \   |")
    print("  |  |    |    ;     | o   T   o |    |    ;     | o   T   o |    |    ;     |  |")
    print("  |  |    |     ;    |  .  |  .  |    |     ;    |  .  |  .  |    |     ;    |  |")
    print("  |  |    |     ;    |   . | .   |    |     ;    |   . | .   |    |     ;    |  |")
    print("  |  |    |     ;    |    .|.    |    |     ;    |    .|.    |    |     ;    |  |")
    print("  |  |    |____;_____|     |     |    |____;_____|     |     |    |____;_____|  |  ")
    print("  |  |   /  __ ;  -  |     !     |   /     '() _-|     !     |  /     '() _-  |  |")
    print("  |  |  / __  ()     |  -      - |  /  __--    -|  -      -  | /  __--     -  |  |")
    print("  |  | /        __-- |    _- _   | /        __--|    _- _   | /        __--_  |  |")
    print("  |__|/________________|_________|/________________|_________|/________________|__|")
    print(" /                                                        _ -                      \ ")
    print("/   -_- _ -                  _- _---                             -_-  -_-         \ ")
    print()

##### END TEXT IMAGE FUNCTIONS ######


##### MAIN PROGRAM CALL #####
# We launch the game from the __main__ environment as usual
if __name__ == '__main__':
    player_info = main(player_info)  # This sends the initialized dictionary to the main function

##### END MAIN PROGRAM CALL #####
`````````
